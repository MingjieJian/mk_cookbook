```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# unit in nm
D = 1

def extinction_wav(lam, D):
    if lam > D:
        return 1 
    elif lam <= D:
        return 0.01

extinction_wav_uf = np.frompyfunc(extinction_wav, 2, 1)
x_plot = np.arange(1e-3, 50, 1e-2)
y_plot = np.zeros(x_plot.shape)

D_list = np.random.randn(5000)*5 + 20
# D_list = np.random.rand(500)*20 + 5
for D in D_list:
    y_plot = y_plot + extinction_wav_uf(x_plot, D)
y_plot = y_plot.astype(float) / len(D_list)
plt.plot(x_plot, y_plot, zorder=1)
plt.gca().twinx()
plt.hist(D_list, color='C1', alpha=0.25, bins=100);
```
